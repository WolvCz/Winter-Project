# Scraping TikTok

```python
tiktok = pd.read_excel('podcast_tiktok.xlsx')
tiktok
```

```python
refined = tiktok[tiktok['URL Type'] == 'Other/Personal']
refined
```

```python
email_list = []

for i in refined['Contact URL1']:
    try:
        soup = BeautifulSoup(requests.get(i).content)
        mailtos = soup.select('a[href^=mailto]')
        if len(mailtos) == 1:
            href = mailtos[0]['href']
            if len(href.split(':')) == 2:
                str1, str2 = href.split(':')
                if '@' in str2:
                    email_list.append(str2)
                else:
                    email_list.append(' ')
            else:
                email_list.append(' ')
        elif len(mailtos)>1:
            try:
                email = mailtos[0]
                href = email['href']
                str1, str2 = href.split(':')
                email_list.append(str2)
            except ValueError:
                email_list.append(' ')
        else:
            email_list.append(' ') 
    except requests.exceptions.ConnectionError as e:
        email_list.append(' ')
```

```python
for i in range(len(refined['Contact URL2'])):
    try:
        soup = BeautifulSoup(requests.get(list(refined['Contact URL2'])[i]).content)
        mailtos = soup.select('a[href^=mailto]')
        if email_list[i] != ' ':
            continue
        else:
            if len(mailtos) == 1:
                href = mailtos[0]['href']
                if len(href.split(':')) == 2:
                    str1, str2 = href.split(':')
                    if '@' in str2:
                        email_list[i] = str2
            elif len(mailtos)>1:
                try:
                    email = mailtos[0]
                    href = email['href']
                    str1, str2 = href.split(':')
                    email_list[i] = str2
                except ValueError:
                    continue
            else:
                continue
    except requests.exceptions.ConnectionError as e:
        continue
```
